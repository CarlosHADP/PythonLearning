```python
class Animal:
    
    alive = True
   
    def eat(self):
        print("This animal is eating")
    
    def sleep(self):
        print('This animal is sleeping')

class Rabbit(Animal):
    def run(self):
        print('This rabbit is running')

class Fish(Animal):
    def swim(self):
        print('This fish is swimming')
        
class Hawk(Animal):
    def fly(self):
        print('This hawk is flying')
    

rabbit = Rabbit()
fish = Fish()
hawk = Hawk()

rabbit.sleep()
print(rabbit.alive)  
rabbit.run()
fish.swim()
hawk.fly()
```

```python
# multi-level inheritance = when a derived (child) class inherits another derived (child) class

class Organism:
    
    alive = True
    
class Animal(Organism):  # Animal is a subclass of Organism, so it inherits from its superclass.
    
    def eat(self):
        print('This animal is eating')
        
class Dog(Animal): #  Dog is also a subclass of Organism, but it's a further subclass of Animal. So it inherits not only from Organism, but from Animal too.    
    def bark(self):
        print('This dog is barking')
        
dog = Dog()

print(dog.alive)
dog.eat()
dog.bark()
```

```python
# multiple inheriterance = when a 'child' class is derived  from more than one parent class

class Prey:
    
    def flee(self):
        print('This Animal flees')
        
class Predator:
    
    def hunt(self):
        print('This animal is hunting')
        
class Rabbit(Prey):
    pass
class Hawk(Predator):
    pass

class Fish(Prey, Predator):
    pass

rabbit = Rabbit()
hawk = Hawk()
fish = Fish()

rabbit.flee()
hawk.hunt()
fish.hunt()
fish.flee()
```
