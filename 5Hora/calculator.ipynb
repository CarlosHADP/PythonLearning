```python
#CALCULATOR 
import math
def calculator():
     while True:
         print("Select operation.")
         print("1.Addition")
         print("2.Subtraction")
         print("3.Multiplication")
         print("4.Division")
         
         try:
             choice = int(input('Chose a number'))
             
             if (choice < 1) or (choice > 4):
                 raise ValueError
         except ValueError:
             print("Invalid Input")
             continue
         
         num1 = float(input("Enter first number : "))
         num2 = float(input("Enter second number :"))
         
         if choice == 1:
             print('Answer:',num1 + num2)
             break
         elif choice == 2:
             print('Answer:',num1 - num2)
             break
         elif choice == 3:
             print('Answer:',num1 * num2)
             break
             # Check for division by zero error
         elif choice == 4:
            print('Answer:',num1 / num2)
            if num2 == 0:
                print("Error! Cannot divide by zero.")
            else:
                print('Answer:',num1 / num2)
                break
               
def play_again():
    again = input("Do you want to perform another calculation? Y/N: ").upper()
    if again=='Y':
        calculator()
    else:
        print('Byee!')
            
calculator()                     
play_again()



# def calculate(x,y): 
#     multiply = x * y
#     divide = x / y
#     add = x + y
#     sub = x - y
#     solve = print(f'Solution= \n Add= {add} \n Sub= {sub} \n Multiply= {multiply} \n Divide= {divide}' )
#     return solve
# def multiply(x,y):
#     print(f'You want to multiply{x} and {y}, here is your solution')
#     multiply = x * y
#     solve = print('Solution: ',multiply)
#     return solve

# def divide(x,y):
#     print(f'You want to divide{x} and {y}, here is your solution')
#     divide = x / y
#     solve = print('Solution: ',divide)
#     return solve

# def add(x,y):
#     print(f'You want to add{x} and {y}, here is your solution')
#     add = x + y
#     solve = print('Solution: ',add)
#     return solve

# def sub(x,y):
#     print(f'You want to subtract {x} and {y}, here is your solution: ')
#     sub = x - y
#     solve = print('Solution: ',sub)
#     return solve




# calculate(5,5)
# # multiply(5,5)
# # add(5,5)
# # divide(5,5)   
# # sub(5,5)
```
